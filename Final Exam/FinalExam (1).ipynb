```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
import warnings
warnings.filterwarnings('ignore')


if __name__ == "__main__":
    
    x = np.arange(0,4,0.1)
    y1 = []
    y2 = []
    y3 = []
 
    iterations = [10, 100, 1000]
    
    
    for it in iterations:

        for i in x:
            summ = 0
            product = 1

            for j in range(1,it+1):
                if(j % 2 != 0):
                    summ = summ + ((np.tan(i*pi))**product)/product
                else:
                    summ = summ - ((np.tan(i*pi))**product)/product

                product += 2

            if(it == iterations[0]):
                y1.append(summ)
            elif(it == iterations[1]):
                y2.append(summ)
            elif(it == iterations[2]):
                y3.append(summ)

    
    
    plt.figure(1)
    plt.plot(x, y1)
    plt.scatter(x, y1)
    plt.ylabel('sum for iterations 10')
    plt.xlabel('x')
    plt.title("Figure 1: Madhava Series for 10 Iterations")
    
   
    plt.figure(2)
    plt.plot(x, y2)
    plt.scatter(x, y2)
    plt.ylabel('sum for iterations 100')
    plt.xlabel('x')
    plt.title("Figure 2: Madhava Series for 100 Iterations")
    
    
    plt.figure(3)
    plt.plot(x, y3)
    plt.scatter(x, y3)
    plt.ylabel('sum for iterations 1000')
    plt.xlabel('x')
    plt.title("Figure 3: Madhava Series for 1000 Iterations")
    
    
   
    
```

## Description of the Figures:

Madhava series: tan x - tan^3 x/3 + tan^5 x/5 - tan^7 x/7 + tan^9 x/9 ..... 

Here, the number of iterations used are 10, 100, 1000. And the range of x values is 0, 0.1pi, 0.2pi, 0.3pi, ......, 4pi.

From the graphs, we can see that, for small number of iterations, the value of y is almost equal to the value of x. But as we increase the number of iterations, the discontinuty in the graph increases. 

As we can see, in Figure 1, discontinuity occurs after the 5 values of x and the discontinuity continues for only 1 single point. But as the number of iterations increases (Fgiure 2 and Figure 3), the frquency and length of discontinuation increases. 

Figure 2 and Figure 3 basically generate a curve similar to 'tagent curve'. Like 'tangent' curve, this curve is also discontinuous at the odd mulple postions of $\Pi$/2. If we increase the number of iterations for Figure 1, that will also look like 'tangent curve'.

For even number of iterations, the curves goes from infinite positive value to infinite negative value. For odd number off iterations (such as 9, 101, 1001), the curves will go from infinite negative value to infinite positive value. 

So, we can draw the conclusion that, for large number of iterations, Madhava series generates a curve similar to the 'tangent' curve.
